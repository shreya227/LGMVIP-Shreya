In This Task: We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image.

#### Importing The required Libraries

```python
import cv2 
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
os.chdir('C:\\Users\\Shreya\\Downloads')
```

#### reading and displaying the image in RBG format

```python
img1 = cv2.imread('Krishna.jpg')
cv2.imshow('Original Image' , img1);
cv2.waitKey(0)                 #to display the image until you press any key in keyboard
cv2.destroyAllWindows()
```

#### Converting the Original Image into Grayscale

```python
gray_img = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
#grey_img = cv2.imread('Krishna.jpg' ,0)
cv2.imshow('Gray-Scale Image' ,gray_img);
cv2.waitKey(0)                                         #to display the image until you press any key in keyboard
cv2.destroyAllWindows()
```

#### Inverting the Grayscale image. Also called as Negative image

```python
invert_img = cv2.bitwise_not(gray_img)
cv2.imshow('Inverted Gray-scale Image',invert_img)
cv2.waitKey(0)                 #to display the image until you press any key in keyboard
cv2.destroyAllWindows()
```

#### To blur the image

```python
# Blur the image
smooth_img = cv2.GaussianBlur(invert_img,(23,23),sigmaX=0, sigmaY=0)
cv2.imshow('Inverted Blurry Image' , smooth_img );
cv2.waitKey(0)                 #to display the image until you press any key in keyboard
cv2.destroyAllWindows()
```

#### Mixing the grayscale image and inverted blur image to get final pencil sketch

```python
final_img = cv2.divide(gray_img, 255-smooth_img, scale=250)
cv2.imshow('Final Sketch' , final_img );
cv2.waitKey(0)                 #to display the image until you press any key in keyboard
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(7,7))
img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.title('original image')

plt.figure(figsize=(7,7))
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.title('Grayscale image')

plt.figure(figsize=(7,7))
plt.imshow(invert_img,cmap='gray')
plt.axis("off")
plt.title('Negative image')

plt.figure(figsize=(7,7))
plt.imshow(smooth_img, cmap='gray')
plt.axis("off")
plt.title('Blurred image')

plt.figure(figsize=(7,7))
plt.imshow(final_img, cmap='gray')
plt.axis("off")
plt.title('Pencil Sketch image')
plt.show()
```
